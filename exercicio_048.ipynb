Esse programa retorna a soma dos números ímpares que são múltiplos de três no intervalo de 1 até 500.

```python
count = 0
soma = 0
for i in range(1, 500, 2):
  if i % 3 == 0:
    count = count + 1
    soma = soma + i
print('Existem {} números que são ímpares e múltiplos de três no intervalo de 1 até 500,\ne a soma desses números é igual a {}'.format(count, soma))
```
