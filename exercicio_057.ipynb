Esse program ler o sexo de um indivíduo e só aceita as opções M ou F

```python
sexo = input('Qual o sexo (M/F): ').upper()

while sexo != 'M' and sexo != 'F':
  print('Valor inválido! Entre com uma resposta válida')
  sexo = input('Qual o sexo (M/F): ').upper()

print('Você escolheu a opção {}.'.format(sexo))
```
