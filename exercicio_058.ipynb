Esse programa é um jogo onde o jogador tenta adivinhar um número entre 0 e 10.

```python
from random import randint
valor = randint(0, 10)

nome_jogo = 'JOGO DE ADIVINHAÇÃO'

print(10*'=' + len(nome_jogo)*'=' + 10*'=')
print(10*'=' + nome_jogo + 10*'=')
print(10*'=' + len(nome_jogo)*'=' + 10*'=')
print('')

chute = int(input('Um número inteiro entre 0 e 10 foi sortiado. Tente adivinha-lo: '))

while chute != valor:
  
  if chute < valor:
    chute = int(input('Você errou, o valor sorteado é maior. Tente novamente: '))
  elif chute > valor:
    chute = int(input('Você errou, o valor sorteado é menor. Tente novamente: '))
  
  if chute == valor:
    print('PARABÉNS!!! VOCÊ ACERTOU.')
```
