Esse program destrincha um número inteiro em unidades dezena e milhar.

```python
n = int(input('Entre com um número inteiro: '))

def num_decomp(n, i, posicao):
    aux = str(n/i)
    aux = int(aux.split('.')[posicao])
    return aux

unidade = num_decomp(n, 10, 1)

aux = num_decomp(n, 100, 1)
dezena = num_decomp(aux, 10, 0)

aux = num_decomp(n, 1000, 1)
centena = num_decomp(aux, 100, 0)

milhar = num_decomp(n, 1000, 0)

print('O número {} com posto por {} unidades, {} dezenas, {} centenas e {} milhares'.format(n, unidade, dezena, centena, milhar))
```
