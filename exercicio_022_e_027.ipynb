Esse program ler o nome completo da pessoa e retorna o seu primeiro e último nome.

```python
nome = input('Qual o seu nome completo: ')

# TRATAMENTO DA STRING
nome = nome.strip()
nome = nome.capitalize()

# SEPARANDO CADA PARTE DO NOME
sep = nome.split()

print('Seu primeiro nome é {} e seu último nome é {}'.format(sep[0], sep[len(sep)-1]))
```
