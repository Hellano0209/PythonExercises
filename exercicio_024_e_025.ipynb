Esse programa ler o nome de uma cidade e verifica se a o nome começa com "Santo".

```python
cidade = input('Qual o nome da cidade que você mora: ').strip()

cidade = cidade.split(' ')

for i in range(len(cidade)):
  cidade[i] = cidade[i].capitalize()

if 'Santo' in cidade:
  print('Essa cidade contém a palavra {}'.format('Santo'))
else:
  print('Essa cidade não contém a palavra {}'.format('Santo'))
```
