Esse programa ler o ano de nascimento de um homem e verifica quanto tempo falta para o alistamento militar.

```python
from datetime import date

ano = int(input('Qual o ano do seu nascimento: '))
hoje = date.today().year

idade = hoje - ano

if idade <= 18:
  falta = 18 - idade
  print('Fata(m) {} ano(s) para você se alistar.'.format(falta))

else:
  falta = 18 - idade
  print('Você deveria ter se alistado em {}, há {} ano(s) atrás.'.format(hoje + falta, -falta))
```
