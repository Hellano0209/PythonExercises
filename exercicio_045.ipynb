```python
from random import randint
from time import sleep

jogo = ' PEDRA - PAPEL - TESOURA '
print((20+len(jogo))*'=')
print(10*'=' + jogo + 10*'=')
print((20+len(jogo))*'=')

print('Suas opções:\n[0]Pedra\n[1]Papel\n[2]Tesoura')
jogador = int(input('Qual a sua jogada: '))

machine = randint(0, 2)
jogadas = ['Pedra', 'Papel', 'Tesoura']

print('')
sleep(0.5)
print('JO')
sleep(0.5)
print('KEN')
sleep(0.5)
print('PO!')
sleep(0.5)
print('')

print(25*'=')
print('Computador jogou {}\nVocê jogou {}'.format(jogadas[machine], jogadas[jogador]))
print(25*'=')

if machine == 0:
  if jogador == 1:
    print('Parabéns você venceu!')
  elif jogador == 0:
    print('O jogo empatou')
  else:
    print('O computador venceu')
elif machine == 1:
  if jogador == 1:
    print('O jogo empatou')
  elif jogador == 0:
    print('O computador venceu')
  else:
    print('Parabéns você venceu!')
else:
  if jogador == 1:
    print('O computador venceu')
  elif jogador == 0:
    print('Parabéns você venceu!')
  else:
    print('O jogo empatou')
```
