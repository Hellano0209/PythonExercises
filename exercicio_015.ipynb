Esse programa calcula o valor a pagar de um aluguel de carro dado a quantidade de dias que o cliente ficou com o carro e a quilometragem percorrida.



1.   Diária do carro custa 60,00R\$
2.   0,15R\$ por quilometro rodado



```python
dias = int(input('Quanto dias você ficou com o veículo: '))
km = float(input('Quilometragem percorrida: '))

valor = 60*dias + 0.15*km

print('Valor do aluguel do veículo é de: R${}.'.format(valor))
```
