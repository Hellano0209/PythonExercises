Esse program ler um angulo e calcula o seno, cosseno e a tangente.

```python
import math

angulo = float(input('Digite um angulo: '))
radiano = math.radians(angulo)

seno = math.sin(radiano)
coss = math.cos(radiano)
tang = math.tan(radiano)

print('O Seno, Cosseno e a Tangente do algulo {} é {:.2f}, {:.2f} e {:.2f}, respectivamente.'.format(angulo, seno, coss, tang))
```
