Esse programa ler 3 números e verificar qual o maior e o menor.

```python
import pandas as pd

num = []
for i in range(3):
  n = float(input('Entre com o {}º número: '.format(i+1)))
  num.append(n)

num.sort()

print('O maior número é {} e o menor é {}'.format(num[2], num[0]))
```
