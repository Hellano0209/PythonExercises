E programa ler o peso de um grupo de pessoas e identifica qual tem o menor e o maior peso.

```python
import pandas as pd

n = int(input('Quantas pessoas: '))

nomes = []
pesos = []
for i in range(n):
  nome = input('Qual o nome da {}ª pessoa: '.format(i+1))
  peso = input('Qual o peso do(a) {}: '.format(nome))

  nomes.append(nome)
  pesos.append(peso)

data = {'nomes':nomes, 'pesos':pesos}
data = pd.DataFrame(data)

pesoMax = data['pesos'].max()
pesoMin = data['pesos'].min()

infoMax = data[['nomes', 'pesos']][data['pesos'] == pesoMax]
infoMin = data[['nomes', 'pesos']][data['pesos'] == pesoMin]

nomeMax = list(infoMax['nomes'])[0]
pesoMax = list(infoMax['pesos'])[0]
nomeMin = list(infoMin['nomes'])[0]
pesoMin = list(infoMin['pesos'])[0]

print('{} é a pessoa com o maior peso com {}kg.'.format(nomeMax, pesoMax))
print('{} é a pessoa com o menor peso com {}kg.'.format(nomeMin, pesoMin))
```
