```python
import random as rd

a = rd.randint(0,5)

print('Pensei em número de entre 0 e 5!')
n = int(input('Qual número eu pensei? '))

if n == a:
  print('Que sorte, você acertou!')
else:
  print('Você errou! O número que eu pensei foi {}'.format(a))
```
