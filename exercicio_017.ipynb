Esse programa ler o cateto oposto e adjacente de um triangulo e calcula a hipotenusa.

```python
from math import sqrt

catop = float(input('Digite o cateto oposto: '))
catad = float(input('Digite o cateto adjacente: '))

hip = sqrt(catop**2 + catad**2)


print('A hipotenusa do triângulo é: {:.2f}'.format(hip))
```
