Esse programa calcula a nota média e verifica qual a situação do aluno.

```python
import pandas as pd

n = int(input('Quantas avaliações foram feitas: '))
notas = []

for i in range(n):
  nota = float(input('Qual a nota da {}ª avaliação: '.format(i+1)))
  notas.append(nota)

notas = pd.DataFrame(notas, columns = ['notas'])
media = notas['notas'].mean()

if media < 4:
  print('A média do aluno é {:.2f}.\nSituação: REPROVADO.'.format(media))
elif media >= 7:
  print('A média do aluno é {:.2f}.\nSituação: APROVADO.'.format(media))
else:
  print('A média do aluno é {:.2f}.\nSituação: RECUPERAÇÃO.'.format(media))
```
