Esse programa ler o comprimento de três segumentos de retas e diz se elas podem formar um triângulo. Caso seja possível formar o triângo o programa também retorno que tipo de triângulo.

Para três segumentos de retas formarem um triângulo, cada segumento deve ter um comprimento menor que a soma dos outros dois.

```python
reta = []

for i in range(3):
  r = float(input('Qual o comprimento do {}º segumento de reta: '.format(i+1)))
  reta.append(r)

cond1 = reta[0] < reta[1] + reta[2]
cond2 = reta[1] < reta[0] + reta[2]
cond3 = reta[2] < reta[1] + reta[0]

if cond1 and cond2 and cond3:
  if reta[0] == reta[1] and reta[0] == reta[2]:
    print('Esses seguimentos formam um triângulo EQUILÁTERO.')
  elif reta[0] != reta[1] and reta[0] != reta[2]:
    print('Esses seguimentos formam um triângulo ESCALENO.')
  else:
    print('Esses seguimentos formam um triângulo ISÓSCELES.')
else:
  print('Esses seguimento não formam um triângulo.')
```
