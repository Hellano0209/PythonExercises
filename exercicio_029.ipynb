Esse program ler a velocidade de um veículo e se a velocidade ultrapassar 80km/h aplica uma multa de R$7,00 a cada quilômetro que excedeu o limite.

```python
km = float(input('Qual a velocidade do veículo: '))

if km > 80:
  aux = km -80
  multa = aux*7
  print('Você excedeu o limite de velocidade em {}km/h. Você pagará uma multa de R${}.'.format(aux, multa))
else:
  print('Você está dentro do limite de velocidade.') 
```
