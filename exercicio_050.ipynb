Esse programa ler seis números e calcula a soma dos números pares.

```python
numeros = []
for i in range(6):
  n = int(input('Digite o {}º número: '.format(i+1)))
  if n % 2 == 0:
    numeros.append(n)

print('A soma dos números pares é {}'.format(sum(numeros)))
```
