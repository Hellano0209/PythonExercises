Esse programa ler um número inteiro e diz se ele é um número primo

```python
n = int(input('Digite um número inteiro: '))

div = []
for i in range(1,n+1):
  if n % i == 0:
    div.append(i)

if len(div) == 2:
  print('O número {} é primo.'.format(n))
else:
  print('O número {} não é primo pois é divisível pelos números {}.'.format(n, div))
```
