Esse programa escreve os 10 primeiros termos de uma Progressão Aritimética.

```python
inic = int(input('Qual o primeiro termo da P.A.: '))
razao = int(input('Qual a razão da P.A.: '))

pa = []
pa.append(inic)
for i in range(9):
  inic = inic + razao
  pa.append(inic)

print('Os 10 primeiros termos da P.A. são: {}'.format(pa))
```
