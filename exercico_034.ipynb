Esse programa calcula o aumento salarial: quem canha até R$1250 recebe um aumento 15% e quem ganha acima desse valor recebe um aumento de 10%.

```python
salario = float(input('Qual o seu salário: '))

if salario <= 1250:
  print('Com um aumento de 15% seu salário passa a ser R${}.'.format(salario*(1+0.15)))
else:
  print('Com um aumento de 10% seu salário passa a ser R${}.'.format(salario*(1+0.10)))
```
