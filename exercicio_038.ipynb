Esse programa ler dois números e verifica qual é o maior ou se são iguais.

```python
numeros = []
for i in range(2):
  n = float(input('Qual o {}º número: '.format(i+1)))
  numeros.append(n)

if numeros[0] > numeros[1]:
  print('O primeiro número é maior.')
elif numeros[0] < numeros[1]:
  print('O segundo número é maior.')
else:
  print('Os números são iguais.')

```
