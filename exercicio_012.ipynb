Esse programa calcula quanto é um desconto de 5% no valor de um produto.


```python
valor = float(input('Qual o valor do produto: '))

taxa = 0.05
desconto = taxa*valor
preco = valor - desconto

print('O valor do produto é {:.2f}R$. Com desconto de 5% fica no valor de {:.2f}R$. O desconto foi de {:.2f}R$!'.format(valor, preco, desconto))
```
