Esse programa ler um número decimal e converte para base binária ou octal ou hexadecimal.

```python
n = int(input('Digite um número inteiro: '))
print('[1] Converter para BINÁRO')
print('[2] Converter para OCTAL')
print('[3] Converter para HEXADECIMAL')

codigo = int(input('Sua opção: '))

if codigo == 1:
  div = 2
elif codigo == 2:
  div = 8
elif codigo == 3:
  div = 16

resto = n % div
inteiro = n // div
convert = ''

while inteiro >= div:
  convert = convert + str(resto)
  resto = inteiro % div
  inteiro = inteiro // div

convert = convert + str(resto) + str(inteiro)

print('O número decimal {} convertido em binário é: {}'.format(n, convert))
```
