Esse programa ler uma fraze e verifica se ela é um palídromo.

```python
frase = input('Digite uma frase: ')

frase = frase.strip().lower()
frase = frase.split(' ')

def join(str1):
  str2 = ''
  for i in range(len(str1)):
    str2 = str2 + str1[i]

  return str2

frase = join(frase)

frase2 = ''
for i in range((len(frase)-1), 0, -1):
  frase2 = frase2 + frase[i]

if frase == frase2:
  print('Frase: {}\nEssa frase é um palídromo.'.format(frase))
else:
  print('Frase: {}\nEssa frase não é um palídromo.'.format(frase))
```
