Esse programa ler uma frase e conta o número de vezes que a letra 'A' aparece e retorna a posição onde ela aparece pela primeira e última vez.

```python
frase = input('Digite uma frase: ').strip().lower()

n = frase.count('a')
p1 = frase.find('a')
pu = frase.rfind('a')

print('A letra \'A\' aparece {} vezes. A primeria vez é na posição {} e última na posição {}.'.format(n + 1, p1 + 1, pu + 1))
```
