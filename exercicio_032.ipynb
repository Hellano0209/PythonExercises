Esse programa identifica se um detarminado ano é bissexto.

```python
from datetime import date

ano = int(input('Qual o ano você que analizar: '))

cond1 = ano % 4 == 0
cond2 = ano % 100 != 0 
cond3 = ano % 400 == 0

if ano <= 0:
  ano = date.today().year

if cond1 and cond2 and cond3:
  print('O ano {} é bissexto.'.format(ano))
else:
  print('O ano {} é não bissexto.'.format(ano))

```
