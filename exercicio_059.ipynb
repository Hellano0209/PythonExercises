Esse programa ler dois número e pergunta qual o tipo de operação o usuário deseja fazer.

```python
import pandas as pd

v1 = int(input('Primeiro valor: '))
v2 = int(input('Segundo valor: '))

opcao = -1

while opcao != 5:
  print('[1] somar')
  print('[2] multiplicar')
  print('[3] maior')
  print('[4] novos números')
  print('[5] sair do programa')

  opcao = int(input('Qual a sua opção: '))

  if opcao == 1:
    print('A soma entre {} e {} é {}.'.format(v1, v2, v1+v2))
    print('')
  elif opcao == 2:
    print('A multiplicação entre {} e {} é {}.'.format(v1, v2, v1*v2))
    print('')
  elif opcao == 3:
    df = pd.DataFrame([v1, v2], columns = ['valores'])
    print('O maior valor entre {} e {} é {}.'.format(v1, v2, df['valores'].max()))
    print('')
  elif opcao == 4:
    v1 = int(input('Primeiro valor: '))
    v2 = int(input('Segundo valor: '))
    print('')
  elif opcao == 5:
    print('Programa finalizado.')

```
