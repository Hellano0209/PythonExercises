Esse progrma concede um emprestimo para financiamento de uma casa com base no valor da casa e no salário do comprador.

```python
casa = float(input('Qual o valor da casa: '))
renda = float(input('Qual a renda do comprador: '))
tempo = int(input('Tempo do financiamento: '))

qtd = tempo*12    # Quantidade prestações
prest = casa/qtd  # Valor da prestação
frenda = prest/renda # Fração da renda que a prestação ocupa

if frenda > 0.3:
  print('A prestação ocupa {:.2f}% da renda do comprador. O seu emprétimo foi negado.'.format(100*frenda))
else:
  print('A prestação ocupa {:.2f}% da renda do comprador. O seu emprétimo foi aceito.'.format(100*frenda))
  print('O valor do financiamento é de R${}, parcelado em {} anos com uma prestação mensal de R${:.2f}.'.format(casa, tempo, prest))
```
