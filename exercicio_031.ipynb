Esse programa calcula o preço de uma passagem de acordo com a distância.

```python
km = float(input('Qual a distância da sua viagem em km: '))

if km <= 200:
  valor = km*0.5
else:
  valor = km*0.45

print('O valor da sua passagem para percorrer uma distância de {}km é R${}.'.format(km, valor))
```
