Esse program ler o nome, idade e o sexo de um grupo de pessoas e retorna a média de idade do grupo, quantas mulheres têm menos de 20 anos e qual é o nome do homem mais velho.

```python
import pandas as pd

n = int(input('Existem quantas pessoas no grupo: '))
print('')

nomes = []
idades = []
sexos = []

for i in range(n):
  nome = input('Qual o nome da {}ª pessoa: '.format(i+1)).capitalize()
  nomes.append(nome)
  idade = int(input('Qual a idade do(a) {}: '.format(nome)))
  idades.append(idade)
  sexo = input('Qual o sexo do(a) {} (M/F): '.format(nome)).upper()
  sexos.append(sexo)
  print('')

dados = pd.DataFrame({'nome':nomes, 'idade':idades, 'sexo':sexos})

# IDADE MÉDIA
media_idade = dados['idade'].mean()
print('A idade média do grupo é {}'.format(media_idade))

# MULHERES COM MENOS DE 20 ANOS
select_f = dados.loc[dados['sexo'] == 'F']
select_f = select_f.loc[select_f['idade'] < 20]

if select_f.empty:
  print('Não tem mulhres com menos de 20 anos no grupo.')
else:
  print('O número de mulheres com menos de 20 anos é: {}'.format(list(select_f.shape)[0]))

# HOMEM MAIS VELHO
select_m = dados.loc[dados['sexo'] == 'M']
select_m.loc[select_m['idade'] == select_m['idade'].max()]

if not select_m.empty:
  nome_m = select_m['nome'][0]
  idade_m = select_m['idade'][0]
  print('{} é o homem com a maior idade: {} anos'.format(nome_m, idade_m))
else:
  print('Não tem homens no grupo.')
```
