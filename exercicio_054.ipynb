Esse programa ler o ano de nascimento de um grupo de pessoas conta quantos tem maioridade e quantos são menores.

```python
from datetime import date
import pandas as pd

n = int(input('Quantas pessoas em no grupo: '))

nowdays = date.today().year
age = []
for i in range(n):
  year = int(input('Qual o ano de nacimento da {}ª pessoa: '.format(i+1)))
  x = nowdays - year
  age.append(x)

dados = pd.DataFrame(age, columns = ['age'])
print('O número de pessoas maiores de 18: {}'.format(len(dados['age'][dados['age'] >= 18])))
print('O número de pessoas menores de 18: {}'.format(len(dados['age'][dados['age'] <  18])))
```
